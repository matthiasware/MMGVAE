```python
import torch
```

```python
x = torch.rand((2, 3))
x = torch.Tensor([[1, 2, 3], [4, 5, 6]])
print(x.shape)
x
```

```python
# permutation
torch.einsum('ij -> ji', x)
```

```python
# summation over all
print(torch.einsum('ij ->', x), x.sum())
```

```python
# row sum
torch.einsum('ij -> i', x)
```

```python
# colsum
torch.einsum('ij -> j', x)
```

```python
# matrix vector multiplication
v = torch.Tensor([0.5, 1, 2])
torch.einsum('ij, j -> ij', x, v)
```

```python
# matrix vecotr multiplication
v = torch.Tensor([0.5, 2])
torch.einsum('ij, i -> ij', x, v)
```

```python
# matrix matrix multiplication
m = torch.Tensor([[1,2,3,4], [5,6,7,8]])
torch.einsum('ij, ik -> jk', x, m)
```

```python
x.T@m
```

```python
# Hadamard Product (element-wise multiplication)
torch.einsum('ij,ij->ij', x, x)
```

```python
# dot product
x.T@v

# (2, 3) (2)
torch.einsum('ij, i -> j', x, v)
```

```python
# Outer Product
a = torch.rand((3))
b = torch.rand((5))
torch.einsum('i,j->ij',a,b)
```

```python
# Batched Matrix Multiplication
a = torch.rand((3, 2, 5))
b = torch.rand((3, 5, 3))
# want (3, 2, 3)

torch.einsum('ijk, ikl -> ijl', a, b).shape
```

```python
# matrix diagonal
x = torch.diag(torch.rand(3))
print(x)
torch.einsum('ii->i', x)
```

```python
# matrix trace
torch.einsum('ii->', x)
```
